#### **O projeto deve evoluir passo a passo, atendendo √†s seguintes regras:**

**1¬∫ Boas-vindas:**
* *O programa deve iniciar mostrando uma mensagem de boas-vindas e identifica√ß√£o do sistema.*

**2¬∫ Coleta de Dados:**
* *O sistema deve solicitar ao usu√°rio seu nome, idade e altura.*
* *A idade deve ser armazenada como n√∫mero inteiro (int) e a altura como n√∫mero decimal (float).*

**3¬∫ Processamento das Informa√ß√µes:**

* *idade > 0 e altura > 0.5 ou se for maior de idade*

**5¬™ Verifica√ß√µes Finais (Programa√ß√£oEstruturada)**
* *O sistema deve exibir uma mensagem de permiss√£o de acesso caso o usu√°rio seja maior de idade; caso contr√°rio, informar que ainda n√£o tem permiss√£o.*

**EXTRA: Organiza√ß√£o em Fun√ß√£o Principal**

###**1. Fun√ß√£o de boas-vindas**

* O sistema deve iniciar exibindo uma mensagem de boas-vindas e identifica√ß√£o do programa.*

```python
def mensagem_inicial():
    print("Bem-vindo ao Sistema de Cadastro 1.0!")
    print("=====================================")

mensagem_inicial()
```

###**2. Fun√ß√£o de coleta de dados**

* O sistema deve coletar do usu√°rio as informa√ß√µes b√°sicas (nome, idade e altura).
* A idade deve ser um n√∫mero inteiro e a altura um n√∫mero decimal.

```python
def coletar_dados():
    nome = input("Digite seu nome: ")
    idade = int(input("Digite sua idade: "))
    altura = float(input("Digite sua altura em metros (ex: 1.75): "))
    return nome, idade, altura #carregar
```

```python
# Teste
coletar_dados()
```

###**3. Fun√ß√£o de processamento**

* O sistema deve calcular o ano de nascimento com base na idade informada e no ano atual (2025).
* Tamb√©m deve verificar se a pessoa √© maior de idade e criar uma vari√°vel booleana para armazenar esse resultado.

```python
def processar_dados(nome, idade, altura):
    ano_atual = 2025
    ano_nascimento = ano_atual - idade
    maioridade = idade >= 18
    soma_altura = altura + 0.05
    validacao = (idade > 0 and altura > 0.5) or maioridade

    return {
        "nome": nome,
        "idade": idade,
        "ano_nascimento": ano_nascimento,
        "altura": altura,
        "soma_altura": soma_altura,
        "maioridade": maioridade,
        "validacao": validacao
    }
```

```python
# ==== Chamando a fun√ß√£o ====
# voc√™ precisa passar os 3 valores
resultado = processar_dados("Vini", 32, 1.70)

print(resultado)
```

###**4. Fun√ß√£o de exibi√ß√£o**

* O sistema deve apresentar todas as informa√ß√µes do cadastro em formato organizado, com mensagens claras e leg√≠veis para o usu√°rio.

```python
def exibir_resumo(dados):
    print("\n===== RESUMO DO CADASTRO =====")
    print(f"Nome: {dados['nome']}")
    print(f"Idade: {dados['idade']}")
    print(f"Ano de Nascimento: {dados['ano_nascimento']}")
    print(f"Altura aproximada (+5cm): {dados['soma_altura']:.2f}m")
    print(f"Maior de idade: {dados['maioridade']}")
    print(f"Valida√ß√£o dos dados: {dados['validacao']}")
```

```python
# 1. Coleta os dados
nome = "Vini"
idade = 32
altura = 1.70
# 2. Processa os dados
resultado = processar_dados(nome, idade, altura)
# 3. Exibe o resumo formatado
exibir_resumo(resultado)
```

*Este √© um exemplo de:*
*Coletar ‚Üí Processar ‚Üí Exibir.*
*No seu c√≥digo, basta ligar as fun√ß√µes da forma certa, passando os valores retornados de uma para a outra.*

###**Quest√£o 5 ‚Äî Fun√ß√£o de Verifica√ß√µes**

O sistema deve:

* Usar uma estrutura de decis√£o para verificar se o usu√°rio √© maior de idade e informar se pode acessar √°reas restritas.
* Fazer uma contagem regressiva (usando for).
* Pedir confirma√ß√£o dos dados at√© o usu√°rio digitar "sim" (usando while).

```python
def verificacoes(dados):
    # 1. Estrutura de decis√£o (if/else)
    if dados["maioridade"]:
        print(f"{dados['nome']}, voc√™ pode acessar √°reas restritas.")
    else:
        print(f"{dados['nome']}, voc√™ ainda n√£o tem permiss√£o total.")

    # 2. Estrutura de repeti√ß√£o (for)
    print("\nContagem regressiva para fim do cadastro:")
    for i in range(3, 0, -1):
        print(i)

    # 3. Estrutura de repeti√ß√£o (while)
    confirmado = ""
    while confirmado.lower() != "sim":
        confirmado = input("Confirma que os dados est√£o corretos? (sim/n√£o): ")

    print("Cadastro conclu√≠do com sucesso!")
```

```python
# 1. Coletar dados
nome, idade, altura = coletar_dados()

# 2. Processar dados
resultado = processar_dados(nome, idade, altura)

# 3. Exibir resumo formatado
exibir_resumo(resultado)

# 4. Aplicar verifica√ß√µes (decis√£o e repeti√ß√£o)
verificacoes(resultado)
```

###**Quest√£o 6 ‚Äî Fun√ß√£o Principal (Fluxo de Processos do Programa)**

O sistema deve centralizar toda a execu√ß√£o chamando, na ordem correta:

* A mensagem inicial.
* A coleta de dados.
* O processamento das informa√ß√µes.
* A exibi√ß√£o dos resultados.
* As verifica√ß√µes finais.



```python
def main():
    # 1. Mensagem inicial
    mensagem_inicial()

    # 2. Coleta de dados
    nome, idade, altura = coletar_dados()

    # 3. Processamento
    dados = processar_dados(nome, idade, altura)

    # 4. Exibi√ß√£o
    exibir_resumo(dados)

    # 5. Verifica√ß√µes (decis√£o + repeti√ß√£o)
    verificacoes(dados)


# Executa apenas se este arquivo for o principal
if __name__ == "__main__":
    main()
```
